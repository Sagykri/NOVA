```python
import os
import shutil
```

```python
# Define the root path
root_path = "/home/labs/hornsteinlab/Collaboration/FUNOVA/input/images/raw"

# Define the cell lines mapping (replace '_' with '-')
cell_lines_mapping = {
    "C9orf72-HRE_1008566": "C9orf72-HRE-1008566",
    "C9orf72-HRE_981344": "C9orf72-HRE-981344",
    "Control_1001733": "Control-1001733",
    "Control_1017118": "Control-1017118",
    "Control_1025045": "Control-1025045",
    "Control_1048087": "Control-1048087",
    "TDP--43-G348V_1057052": "TDP--43-G348V-1057052",
    "TDP--43-N390D_1005373": "TDP--43-N390D-1005373",
}
```

```python
# Walk through the directory structure
for dirpath, dirnames, filenames in os.walk(root_path, topdown=False):  # Use topdown=False for renaming
    # Process files
    for filename in filenames:
        for old_name, new_name in cell_lines_mapping.items():
            if old_name in filename:
                # Replace old cell line name with new one in the file name
                new_filename = filename.replace(old_name, new_name)
                # Rename the file
                old_filepath = os.path.join(dirpath, filename)
                new_filepath = os.path.join(dirpath, new_filename)
                shutil.move(old_filepath, new_filepath)
#                 print(f"Renamed file: {old_filepath} -> {new_filepath}")

    # Process directories
    for dirname in dirnames:
        for old_name, new_name in cell_lines_mapping.items():
            if old_name in dirname:
                # Replace old cell line name with new one in the directory name
                new_dirname = dirname.replace(old_name, new_name)
                # Rename the directory
                old_dirpath = os.path.join(dirpath, dirname)
                new_dirpath = os.path.join(dirpath, new_dirname)
                shutil.move(old_dirpath, new_dirpath)
                print(f"Renamed directory: {old_dirpath} -> {new_dirpath}")
```
