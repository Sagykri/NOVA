```python
import pandas as pd
```

```python
def add_id_to_df(df):
    df['ID'] = df['batch'].astype(str) + '/' + df['cell_line'].astype(str) + '/' + df['panel'].astype(str) + '/' +  df['condition'].astype(str) + '/' +  df['rep'].astype(str) + '/' + df['marker'].astype(str) + '/' + df['filename'].astype(str)
    return df
```

```python
df_old = pd.read_csv("/home/projects/hornsteinlab/Collaboration/MOmaps/outputs/preprocessing/spd/logs/preprocessing_FUS/batch1/original_all_good_but_specfici_markers/cell_count_stats_230224_112751_143123.csv")
df_old = add_id_to_df(df_old)
df_old
```

```python
df_new = pd.read_csv("/home/projects/hornsteinlab/Collaboration/MOmaps/outputs/preprocessing/spd/logs/preprocessing_FUS/batch1/cell_count_stats_260224_202119_247477.csv")#cell_count_stats_250224_172417_280703.csv")
df_new = df_new[df_new['marker']!='DAPI'] # Drop DAPI otherwise there are duplicates since DAPI can come from different filenames
df_new = add_id_to_df(df_new)
df_new
```

```python
df_final = df_old.copy()
# drop old values
df_final = df_old[~df_old['ID'].isin(df_new['ID'])]
# append new ones
df_final = pd.concat([df_final, df_new], ignore_index=True)
df_final.drop(columns=['ID'], inplace=True)
df_final
```

```python
# Save final to file
df_final.to_csv("/home/projects/hornsteinlab/Collaboration/MOmaps/outputs/preprocessing/spd/logs/preprocessing_FUS/merged/batch1/cell_count_stats_270224.csv", index=False)
```
