```python
from sklearn.preprocessing import label_binarize

# Extract class probability columns
class_columns = df.columns.difference(['true', 'predicted', 'correct', 'confidence'])
y_true = df['true'].values
y_score = df[class_columns].values

# Ensure same order as columns
class_names = list(class_columns)
y_true_bin = label_binarize(y_true, classes=class_names)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
for i, class_name in enumerate(class_names):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_score[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{class_name} (AUC = {roc_auc:.2f})")

plt.plot([0, 1], [0, 1], 'k--', label='Chance')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("One-vs-Rest ROC Curves")
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt

# 1. Merge FUSHeterozygous_Untreated + FUSHomozygous_Untreated into FUS_Mut
df['true_merged'] = df['true'].replace({
    'FUSHeterozygous_Untreated': 'FUS_Mut',
    'FUSHomozygous_Untreated': 'FUS_Mut'
})
df['predicted_merged'] = df['predicted'].replace({
    'FUSHeterozygous_Untreated': 'FUS_Mut',
    'FUSHomozygous_Untreated': 'FUS_Mut'
})

# 2. Get label order for consistent display
labels_order = sorted(df['true_merged'].unique())

# 3. Confusion matrix
cm = confusion_matrix(df['true_merged'], df['predicted_merged'], labels=labels_order)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_order)
disp.plot(xticks_rotation=90)
plt.title("Confusion Matrix with FUS_Mut Grouped")
plt.tight_layout()
plt.show()

# 4. Classification report
print("=== Classification Report with FUS_Mut Grouped ===")
print(classification_report(df['true_merged'], df['predicted_merged']))
```

```python
FUS_MUT = ['FUSHeterozygous_Untreated', 'FUSHomozygous_Untreated']
ALS_MUT = ['OPTN_Untreated', 'TBK1_Untreated', 'TDP43_Untreated']
WT_LIKE = ['WT_Untreated', 'FUSRevertant_Untreated']
def predict_group(row):
    group_scores = {
        'FUS_Mut': row[FUS_MUT].sum(),
        'ALS_Mut': row[ALS_MUT].sum(),
        'WT_Like': row[WT_LIKE].sum()
    }
    return max(group_scores, key=group_scores.get)

df['grouped_true'] = df['true'].map({
    **{k: 'FUS_Mut' for k in FUS_MUT},
    **{k: 'ALS_Mut' for k in ALS_MUT},
    **{k: 'WT_Like' for k in WT_LIKE}
})

df['grouped_pred'] = df.apply(predict_group, axis=1)
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

print(classification_report(df['grouped_true'], df['grouped_pred']))

cm = confusion_matrix(df['grouped_true'], df['grouped_pred'], labels=['FUS_Mut', 'ALS_Mut', 'WT_Like'])
ConfusionMatrixDisplay(cm, display_labels=['FUS_Mut', 'ALS_Mut', 'WT_Like']).plot()
plt.title("Rule-Based Group Classifier (Max Sum Prob)")
plt.show()
```

```python
# Only keep original model output probabilities
prob_cols = [col for col in df.columns if col not in ['true', 'predicted', 'correct', 'confidence', 'grouped_true', 'grouped_pred'] and df[col].dtype in [float, 'float64', 'float32']]

# Melt into long-form
df_long = df.melt(id_vars=['true'], value_vars=prob_cols, var_name='predicted_label', value_name='probability')


# Optional: restrict to specific true labels
# df_long = df_long[df_long['true'].isin(['WT_Untreated', 'FUSRevertant_Untreated'])]

# Plot
g = sns.catplot(
    data=df_long, kind='violin',
    x='predicted_label', y='probability',
    col='true', col_wrap=1,  # one row per true label
    height=3.5, aspect=3, cut=0, density_norm='width'
)

g.set_titles("{col_name}")
for ax in g.axes.flatten():
    ax.tick_params(axis='x', rotation=90)

# Adjust spacing to avoid clipping
g.fig.subplots_adjust(top=0.95, hspace=0.4)
g.fig.suptitle("Per-True-Label Predicted Probability Distributions", fontsize=16)

plt.show()
```

```python
grouped = df_long.groupby(['true', 'predicted_label'])['probability'].mean().unstack()

plt.figure(figsize=(10, 5))
sns.heatmap(grouped, cmap='viridis', annot=True, fmt=".2f")
plt.title("Average Predicted Probability per True Label")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()
```

```python
true_labels = df_long['true'].unique()
predicted_labels = df_long['predicted_label'].unique()

fig, axes = plt.subplots(
    len(true_labels), len(predicted_labels),
    figsize=(2.5 * len(predicted_labels), 2 * len(true_labels)),
    sharex=True, sharey=True
)

for i, true in enumerate(true_labels):
    for j, pred in enumerate(predicted_labels):
        ax = axes[i, j]
        subset = df_long[(df_long['true'] == true) & (df_long['predicted_label'] == pred)]
        color = 'darkorange' if true == pred else 'steelblue'
        sns.histplot(
            subset['probability'],
            bins=25,
            ax=ax,
            kde=False,
            color=color
        )
        ax.grid(True, linestyle='--', alpha=0.3)

        # Axis labels
        if i == len(true_labels) - 1:
            ax.set_xlabel(pred, rotation=90, fontsize=9)
        else:
            ax.set_xlabel('')
            ax.set_xticks([])

        if j == 0:
            ax.set_ylabel(true, fontsize=9)
        else:
            ax.set_ylabel('')
            ax.set_yticks([])

        ax.tick_params(axis='both', which='major', labelsize=7)

# Tighten layout
plt.suptitle("Predicted Probability Histograms (True  Predicted)", fontsize=16)
plt.subplots_adjust(wspace=0.1, hspace=0.3, top=0.94)
plt.show()
```
